```python
import pandas as pd
import numpy as np
```

```python
# Load data with right types.
dtypes = {
    'Datum': str,
    'Wochentag': str,
    'Linienverlauf': str,
    'SollAb': str,
    'GPS-Breite Soll': np.float64,
    'GPS-Länge Soll': np.float64,
}

df_linie1_hin_1 = pd.read_csv(filepath_or_buffer='Echtzeitdaten_Linie1_hin_sheet1.csv', sep=';', dtype=dtypes, thousands=',')
df_linie1_hin_2 = pd.read_csv(filepath_or_buffer='Echtzeitdaten_Linie1_hin_sheet2.csv', sep=';', dtype=dtypes, thousands=',')
df_linie1_hin_3 = pd.read_csv(filepath_or_buffer='Echtzeitdaten_Linie1_hin_sheet3.csv', sep=';', dtype=dtypes, thousands=',')
df_linie1_hin_4 = pd.read_csv(filepath_or_buffer='Echtzeitdaten_Linie1_hin_sheet4.csv', sep=';', dtype=dtypes, thousands=',')
df_linie1_hin_1
```

```python
df_linie1_hin_1.dtypes
```

```python
df_linie1_hin = pd.concat([df_linie1_hin_1, df_linie1_hin_2, df_linie1_hin_3, df_linie1_hin_4])
df_linie1_hin
```

```python
# Is departure time is missing sometimes.
len(df_linie1_hin[df_linie1_hin['IstAb'] == ' '])
```

```python
# Drop rows with missing values.
df_linie1_hin = df_linie1_hin.drop(index=(df_linie1_hin[df_linie1_hin['IstAb'] == ' ']).index)
len(df_linie1_hin[df_linie1_hin['IstAb'] == ' '])
```

```python
df_linie1_hin.shape
```

```python
# Convert departure times to datetimes while setting the correct date.
df_linie1_hin['SollAb'] = pd.to_datetime(df_linie1_hin['Datum'] + ' ' + df_linie1_hin['SollAb'], format='%d.%m.%Y %H:%M:%S')
df_linie1_hin['IstAb'] = pd.to_datetime(df_linie1_hin['Datum'] + ' ' + df_linie1_hin['IstAb'], format='%d.%m.%Y %H:%M:%S')
df_linie1_hin['Datum'] = pd.to_datetime(df_linie1_hin['Datum'], format='%d.%m.%Y')
df_linie1_hin.dtypes
```

```python
# Discard arrival times as they are not really important.
df_linie1_hin.drop(columns=['SollAn', 'IstAn'], inplace=True)
df_linie1_hin
```

```python
df_linie1_hin.index
```

```python
# Reindex dataframe because the index got messed up when concatenating.
index = [i for i in range(df_linie1_hin.shape[0])]
df_linie1_hin["index"] = pd.Series(index, index=df_linie1_hin.index)
df_linie1_hin.set_index("index", drop=True,  inplace=True)
df_linie1_hin
```

```python
df_linie1_hin.index
```

```python
df_linie1_hin['Verspätung'] = df_linie1_hin['IstAb'] - df_linie1_hin['SollAb']
df_linie1_hin
```

```python
df_linie1_hin['Verspätung_total_seconds'] = df_linie1_hin['Verspätung'].apply(lambda x: x.total_seconds())
```
