```python
import sklearn 
import pandas as pd
import requests
from bs4 import BeautifulSoup
from dateutil.relativedelta import *
import pickle
import numpy as np
```

```python
with open('df_linie1_hin.pickle','rb') as fp:
    df_linie1_hin = pickle.load(fp)


with open('df_holidays.pickle','rb') as fp:
    df_holidays = pickle.load(fp)
    
with open('df_weather.pickle','rb') as fp:
    df_weather = pickle.load(fp)
```

```python
df_weather
```

```python
df_linie1_hin_merged = pd.merge(df_linie1_hin, df_holidays, how="left", left_on="Datum", right_on="date" )
df_linie1_hin_merged
```

```python
df_linie1_hin_merged[df_linie1_hin_merged["date"]=="2017-08-14"]
```

```python
# Binarize holiday feature.
df_linie1_hin_merged["Feiertag"]=df_linie1_hin_merged["date"].apply(lambda x: not(pd.isnull(x)))
```

```python
df_linie1_hin_merged
```

```python
df_linie1_hin_merged.drop(columns=["holiday_region","name","weekday","date"], inplace=True)
df_linie1_hin_merged
```

```python
# Sort for merging.
df_linie1_hin_merged.sort_values('SollAb', inplace=True)
df_weather.sort_values('Stunden', inplace=True)
# Merge weather data by nearest by timepoint.
df_merged = pd.merge_asof(
    left=df_linie1_hin_merged, 
    right=df_weather, 
    left_on='SollAb',
    right_on='Stunden',
    direction='nearest'
)
df_merged
```
