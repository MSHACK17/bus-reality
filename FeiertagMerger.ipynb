```python
import sklearn 
import pandas as pd
import requests
from bs4 import BeautifulSoup
from dateutil.relativedelta import *
import pickle
import numpy as np
```

```python
with open('df_linie1_hin.pickle','rb') as fp:
    df_linie1_hin = pickle.load(fp)


with open('df_holidays_legal.pickle','rb') as fp:
    df_holidays_legal = pickle.load(fp)
    
with open('df_holidays_not_legal.pickle','rb') as fp:
    df_holidays_not_legal = pickle.load(fp)
    
with open('df_bridge_days.pickle','rb') as fp:
    df_bridge_days = pickle.load(fp)
```

```python
df_holidays = pd.concat([df_holidays_legal,df_holidays_not_legal,df_bridge_days])
df_holidays.reset_index()
```

```python

df_linie1_hin_merged = pd.merge(df_linie1_hin, df_holidays, how="left", left_on="Datum", right_on="date" )
```

```python
df_linie1_hin_merged[df_linie1_hin_merged["date"]=="2017-08-14"]
```

```python
df_linie1_hin_merged["Feiertag"]=df_linie1_hin_merged["date"].apply(lambda x: not(pd.isnull(x)))
```
